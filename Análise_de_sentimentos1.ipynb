```python
import matplotlib.pyplot as plt

# Dados de exemplo (tipos de alimentos e quantidades)
sentimentos = ['Tristeza', 'Alegria', 'Medo', 'Nojo', 'Raiva']
quantidades = [14.25, 35.77, 0, 14.25, 14.25]  # Em porcentagem

# Plotar o gráfico de pizza
plt.pie(quantidades, labels=sentimentos, autopct='%1.1f%%', startangle=140)

# Personalização do gráfico
plt.title('Análise de sentimentos em notícia na ABEC')

# Mostrar o gráfico
plt.savefig ("grafiv.png")

#Link do artigo https://www.abecbrasil.org.br/2024/04/abec-brasil-apoia-ajustes-em-chamada-cnpqcapes-para-apoio-aos-periodicos-nacionais/

```
